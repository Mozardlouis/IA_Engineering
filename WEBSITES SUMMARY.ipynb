```python
# Import the libraries

import os
import requests
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
from openai import OpenAI
import ollama

#!ollama pull deepseek-r1:1.5b 
```

![GPT Model Animation](https://media.giphy.com/media/v1.Y2lkPTc5MGI3NjExY2Y3ZG5tcDYzY29pYXFtNmZrYmR5ejY1bnVvM2V4MWdoNHNmZml1MCZlcD12MV9naWZzX3NlYXJjaCZjdD1n/ZVik7pBtu9dNS/giphy.gif)

```python
#Website structure
class website:
    def __init__(self,link):
        self.url=link
        response=requests.get(link,headers=header)
        soup=BeautifulSoup(response.content, 'html.parser')
        self.title=soup.title.string if soup.title else "No title for the website"
        for element in soup.body(["script","style", "img", "input"]):
            element.decompose()
            self.text=soup.body.get_text(separator="\n", strip=True)

```

```python
# Establish the connexion with GPT

load_dotenv(override=True)
secret_key=os.getenv('OPEN_AI_KEY') # there must be only one file saved as .env in the folder and the getenv() function will open it and read the value for <<OPEN_AI_KEY=>>



# We can check the api found , if any. 
if not secret_key : 
    print("There is no API")
elif not secret_key.startswith("sk-proj"): # the OPEN_AI API must start with "sk-proj" otherwhise it's not OPEN_AI's
    print("The API is not OPEN_AI's")
elif secret_key.strip() != secret_key: # no space nor tab character must be found in the api
    print("The API contains space or tab character, they must be removed")
else:
    print("The API is correct")

openai=OpenAI(api_key=secret_key)
```

```python
# We build structure for the Frontier Model Input and Output

header= {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}




# System Prompt
'''
 The system prompt is the prompt(text command) that defines the type, limit, capacity... of the system which will reply
'''

system_prompt = "You are an assistant that analyzes the contents of a website \
and provides a short summary, ignoring text that might be navigation related. \
Respond in markdown."




# User Prompt
'''
The user prompt is the request submitted the to system
'''

def user_prompt(website):
    user_prompt= f" please, provide a short summary for the website {website.title}, summarise the news and announcements as well, the contents are \n\n\n {website.text}"
    return user_prompt






# Messages / Commands : System and User

def message(website):
    return [{"role": "system", "content": system_prompt},
            {"role": "user" , "content": user_prompt(website)}]



def summarize(link):
    webs=website(link)
    response=openai.chat.completions.create(model="gpt-4o-mini",
    messages=message(webs))
    return response.choices[0].message.content


# something more beautiful
def website_summary(link):
    try:
        summary=summarize(link)
        display(Markdown(summary))
    except Exception:
        return f"there is an error on this page: {link}"
```

<span style="color:blue;font-size:28px; font-weight:bold;"> 2.- OLLAMA MODEL</span>

## ü¶ô Ollama Model in Action

Experience the power of running LLMs **locally** with low latency and full control ‚Äî just like Ollama!


![Ollama Local Model Vibe](https://media.giphy.com/media/v1.Y2lkPTc5MGI3NjExczJiaDJreTJ4NjB2bTNuOHBvaXk3dW42YnhjbmI0bTl5bGVvN3FocCZlcD12MV9naWZzX3NlYXJjaCZjdD1n/xT9IgzoKnwFNmISR8I/giphy.gif)


```python
#Website structure
class website:
    def __init__(self,link):
        self.url=link
        response=requests.get(link,headers=HEADERS)
        soup=BeautifulSoup(response.content, 'html.parser')
        self.title=soup.title.string if soup.title else "No title for the website"
        for element in soup.body(["script","style", "img", "input"]):
            element.decompose()
            self.text=soup.body.get_text(separator="\n", strip=True)

```

```python
# Same request only change the IA model GPT to OLLAMA

OLLAMA_API = "http://localhost:11434/api/chat" # a free API
HEADERS = {"Content-Type": "application/json"}

MODEL = "llama3.2"



def user_prompt_ollama(website):
    user_prompt= f" please, provide a short summary for the website **{website.title}**, summarise the news and announcements as well. the contents are \n\n\n {website.text}"
    return user_prompt

def message_ollama(website):
    return [{"role": "user" , "content": user_prompt_ollama(website)}]


def summarize_ollama(link):
    link = website(link)
    payload = {
        "model": MODEL,
        "messages":  message_ollama(link),
        "stream": False}
    response = requests.post(OLLAMA_API, json=payload, headers=HEADERS)
    content=(response.json()['message']['content'])
    # Format the summary into Markdown
    markdown_summary = f"## üì∞ Summary for [{link.title}]({link.url})\n\n" + content.replace("\n", "\n\n")
    return markdown_summary

def display_better_ollama(link):
    summary=summarize_ollama(link)
    display(Markdown(summary))
```

```python
display_better_ollama("https://foxnews.com")
```


<span style="color:blue;font-size:18px; font-weight:bold;">2.2.- OLLAMA using OLLAMA library & OPEN_AI</span>

```python
# Using OLLAMA Library


def user_prompt_ollama2(website):
    user_prompt= f" please, provide a short summary for the website **{website.title}**, summarise the news and announcements as well."
    return (
        f"{user_prompt} \n\n website content:\n\n {website.text}"
    )
    
def message_ollama(website):
    return [{"role": "user", "content": user_prompt_ollama2(website)}]


def directly_ollama(link):
    link=website(link)
    response = ollama.chat(model=MODEL, messages= message_ollama(link))
    content=response['message']['content']
    # Format the summary into Markdown
    markdown_summary = f"## üì∞ Summary for [{link.title}]({link.url})\n\n" + content.replace("\n", "\n\n")
    return markdown_summary

def display_better_ollama2(link):
    summary=directly_ollama(link)
    display(Markdown(summary))


display_better_ollama2("https://foxnews.com")
```

```python
# Using OPEN_AI

def via_open_ai(link):
    link=website(link)
    ollama_via_openai = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
    response = ollama_via_openai.chat.completions.create(
    model=MODEL,
    messages=message_ollama(link))
    content=response.choices[0].message.content
    # Format the summary into Markdown
    markdown_summary = f"## üì∞ Summary for [{link.title}]({link.url})\n\n" + content.replace("\n", "\n\n")
    return markdown_summary


def webs_summary_ollama3(link):
    try:
        summary=via_open_ai(link)
        display(Markdown(summary))
    except Exception as e:
        display(Markdown(f"‚ö†Ô∏è **Error summarizing** `{link}`:\n\n`{e}`"))

webs_summary_ollama3("https://foxnews.com")
```


<span style="color:blue;font-size:28px; font-weight:bold;">3.- DEEPSEEK MODEL</span>


![DeepSeek GIF](https://media.giphy.com/media/v1.Y2lkPTc5MGI3NjExa2NhbG1saXV4enZhdjJoaGNsZWtsMmJ0ZmZpcTVndHBvMTRnbTV1diZlcD12MV9naWZzX3NlYXJjaCZjdD1n/l3vR85PnGsBwu1PFK/giphy.gif)

```python
def deepseek_summary(link):
    link=website(link)
    ollama_via_openai = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
    response = ollama_via_openai.chat.completions.create(
    model="deepseek-r1:1.5b",
    messages=message_ollama(link))
    content=response.choices[0].message.content
    # Format the summary into Markdown
    markdown_summary = f"## üì∞ Summary for [{link.title}]({link.url})\n\n" + content.replace("\n", "\n\n")
    return markdown_summary



def webs_summary_deepseek(link):
    try:
        summary=deepseek_summary(link)
        display(Markdown(summary))
    except Exception as e:
        display(Markdown(f"‚ö†Ô∏è **Error summarizing** `{link}`:\n\n`{e}`"))

webs_summary_deepseek("https://foxnews.com")
```

```python
# --- List of websites ---
urls = [
    "https://foxnews.com",
    "https://cnn.com",
    "https://www.bbc.com/news",
    "https://www.weforum.org/about/world-economic-forum",
    "https://www.scientificamerican.com",
    "https://hbr.org",
    "https://ourworldindata.org"
]

# --- List of summary functions to run ---
summary_functions = [
    website_summary,               # assume it returns a summary
    display_better_ollama,         # Ollama (requests-based)
    display_better_ollama2,        # Ollama (chat method)
    webs_summary_ollama3,          # OpenAI client (local)
    webs_summary_deepseek          # another custom function
]

# --- Loop through each URL and each function ---
for url in urls:
    print(f"\nüîó Processing: {url}")
    for func in summary_functions:
        try:
            print(f"‚Üí Running `{func.__name__}`...")
            result = func(url)
            # Only display if result is not None and not already displayed inside the function
            if result:
                from IPython.display import Markdown, display
                display(Markdown(str(result)))
        except Exception as e:
            print(f"‚ö†Ô∏è Error in `{func.__name__}` for {url}: {e}")
```
